```python
def reduce(polymer):
    flag = 0
    new = polymer
    while flag < 1:
        start_len = len(new)
        for i in range(ord('a'),ord('z')+1):
            little_big = str(chr(i)+chr(i-32))
            big_little = str(chr(i - 32)+chr(i))
            new = new.replace(little_big,'')
            new = new.replace(big_little,'')
        end_len = len(new)
        if start_len == end_len:
            flag += 1
            return(new)
```

```python
results = []
for i in range(ord('a'),ord('z')+1):
        new = polymer.replace(chr(i),'')
        new = new.replace(chr(i-32),'')
       # print(new[:100])
        reduced_list = reduce(new)
        results.append([len(reduced_list),i])
```
