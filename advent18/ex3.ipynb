```python
#feed in two indexts for different rectangles 
def overlap_area(i,j,xys):
#     if xys[j][0] > xys[i][0] and xys[j][0] < xys[i][2]:
#         #print('overlapx')
#         overlapx = xys[i][2] - xys[j][0]
#         if xys[j][1] > xys[i][1] and xys[j][1] < xys[i][3]:
#             #print('overlapy')
#             overlapy = xys[i][3] - xys[j][1]
#             overlaparea = overlapy*overlapx
#         else:
#             overlaparea = 0
    x = max((min(xys[i][2], xys[j][2]) - max(xys[i][0], xys[j][0])),0)
    y = max((min(xys[i][3], xys[j][3]) - max(xys[i][1], xys[j][1])),0)

    overlaparea= x * y 
#     (min(r1.x, r2.x) -  
#                  max(l1.x, l2.x)) *  
#                 (min(r1.y, r2.y) - 
#                  max(l1.y, l2.y))
#     else:
#         overlapx = 0
#         overlapy = 0
#         overlaparea = overlapx*overlapy
    return overlaparea
```
